The problem with the above is that we need to be consistent between the temporal scale of $z$ and the other variables.

```python
## Agent model (simple)

pgm = daft.PGM()

# Instance
pgm.add_node("l", r"$\ell$", 0.5, 2)

# Data.
pgm.add_node("u", r"$u_{t-1}$", 0.5, 0)
pgm.add_node("o", r"$o_{t}$", 1.5, 1)

#Latent
pgm.add_node("z", r"$z_t$", 2.5, 1)

pgm.add_node("pi", r"$\pi$", 3.5, 0)

pgm.add_node("V", r"$\hat{V}^{\ \pi}$", 3.5, 2)

# Add in the edges.
pgm.add_edge("l", "u")
pgm.add_edge("l", "o")
pgm.add_edge("u", "o")
pgm.add_edge("o", "z")
pgm.add_edge("z", "pi")
pgm.add_edge("z", "V")

# Render and save.
pgm.render()
```

```python
## Agent model (full)

pgm = daft.PGM()

# Instance
pgm.add_node("l", r"$\ell$", 0.5, 2)

# Data.
pgm.add_node("s_prev", r"$s_{t-1}$", 0.5, 0)
pgm.add_node("s", r"$s_{t}$", 1.5, 1)
pgm.add_node("o", r"$o_{t}$", 2.5, 1)

#Latent
pgm.add_node("z", r"$z_t$", 3.5, 1)

pgm.add_node("pi", r"$\pi$", 4.5, 0)

pgm.add_node("V", r"$\hat{V}^{\ \pi}$", 4.5, 2)

# Add in the edges.
pgm.add_edge("l", "s_prev")
pgm.add_edge("l", "s")
pgm.add_edge("l", "o")
pgm.add_edge("s_prev", "s")
pgm.add_edge("s", "o")
pgm.add_edge("o", "z")
pgm.add_edge("z", "pi")
pgm.add_edge("z", "V")

# Render and save.
pgm.render()
```
